```python
class NoopProcessor(BasePostprocessor):
    def __init__(self):
        super().__init__(None)
        self.APS_mode = False
        self.hyperparam_search_done = False

    @torch.no_grad()
    def postprocess(self, net: nn.Module, data):
        output = net(data)
        logits, features = net(data, return_feature=True)
        conf, pred = torch.max(logits, dim=1)

        return pred, conf
```

```python



class RFFMeanPostprocessor(BasePostprocessor):
    """
    Random Fourier Feature Mean Embedding OOD detector
    Fully self-contained (no external YAML config needed)
    """

    def __init__(
        self,
        num_rff=4096,
        sigma=0.5,
        fit_ratio=0.95,
    ):
        super().__init__(None)

        # Hyperparameters live inside class
        self.D = num_rff
        self.sigma = sigma
        self.fit_ratio = fit_ratio

        self.setup_flag = False

        self.APS_mode = False
        self.hyperparam_search_done = False


    def _sample_rff_params(self, feat_dim, device):
        self.omega = torch.randn(
            feat_dim, self.D, device=device
        ) / self.sigma

        self.b = 2 * np.pi * torch.rand(self.D, device=device)

    def _phi(self, x):
        """
        x: [N, feat_dim]
        return: [N, D]
        """
        projection = x @ self.omega + self.b
        return torch.sqrt(
            torch.tensor(2.0 / self.D, device=x.device)
        ) * torch.cos(projection)



    def setup(self, net: nn.Module, id_loader_dict, ood_loader_dict):

        if self.setup_flag:
            return

        print("\nEstimating RFF mean embedding...")

        net.eval()
        all_feats = []

        with torch.no_grad():
            for batch in tqdm(id_loader_dict['train'] , desc="calibrating mean"):
                data = batch['data'].cuda()
                _, features = net(data, return_feature=True)
                features = features / torch.norm(features , dim = 1 , keepdim=True)
                all_feats.append(features)

        all_feats = torch.cat(all_feats, dim=0)
        N, feat_dim = all_feats.shape
        device = all_feats.device


        perm = torch.randperm(N, device=device)

        X_fit = all_feats

        self._sample_rff_params(feat_dim, device)

        phi_fit = self._phi(X_fit)
        self.mu_hat = phi_fit.mean(dim=0)  # [D]


        self.setup_flag = True

    @torch.no_grad()
    def postprocess(self, net: nn.Module, data):
        output = net(data)
        logits, features = net(data, return_feature=True)
        features = features / torch.norm(features , dim = 1 , keepdim=True)
        conf, pred = torch.max(logits, dim=1)
        self.phi_x = self._phi(features)
        m_x = self.phi_x @ self.mu_hat
        conf = m_x
        return pred, conf



```

```python
from tqdm import tqdm
import numpy as np
import torch
import torch.nn as nn
from openood.postprocessors import BasePostprocessor


class ClassConditionalRFFMeanPostprocessor(BasePostprocessor):
    """
    Class-Conditional Random Fourier Feature Mean Embedding OOD detector
    """

    def __init__(
        self,
        num_rff=4096,
        sigma=0.5,
        fit_ratio=0.95,
    ):
        super().__init__(None)

        self.D = num_rff
        self.sigma = sigma
        self.fit_ratio = fit_ratio

        self.mu_hat_per_class = []
        self.setup_flag = False

        self.APS_mode = False
        self.hyperparam_search_done = False



    def _sample_rff_params(self, feat_dim, device):
        self.omega = torch.randn(
            feat_dim, self.D, device=device
        ) / self.sigma

        self.b = 2 * np.pi * torch.rand(self.D, device=device)

    def _phi(self, x):
        """
        x: [N, feat_dim]
        return: [N, D]
        """
        projection = x @ self.omega + self.b
        return torch.sqrt(
            torch.tensor(2.0 / self.D, device=x.device)
        ) * torch.cos(projection)



    def setup(self, net: nn.Module, id_loader_dict, ood_loader_dict):

        if self.setup_flag:
            return

        print("\nEstimating Class-Conditional RFF mean embeddings...")

        net.eval()
        num_classes = net.fc.out_features
        self.mu_hat_per_class = [None] * num_classes

        all_feats = []
        all_labels = []

        with torch.no_grad():
            for batch in tqdm(id_loader_dict['train'], desc="calibrating class-conditional mean embeddings"):
                data = batch['data'].cuda()
                labels = batch['label'].cuda()
                _, features = net(data, return_feature=True)
                features = features / torch.norm(features , dim = 1 , keepdim=True)
                all_feats.append(features)
                all_labels.append(labels)

        all_feats = torch.cat(all_feats, dim=0)
        all_labels = torch.cat(all_labels, dim=0)
        N, feat_dim = all_feats.shape
        device = all_feats.device

        self._sample_rff_params(feat_dim, device)
        for c in range(num_classes):
            class_feats = all_feats[all_labels == c]
            if len(class_feats) > 0:
                phi_fit = self._phi(class_feats)
                self.mu_hat_per_class[c] = phi_fit.mean(dim=0)
            else:
                print(f"Warning: No training samples found for class {c}. Skipping mean embedding calculation for this class.")

        self.setup_flag = True
        print("Class-Conditional RFF mean embeddings estimated.")

    @torch.no_grad()
    def postprocess(self, net: nn.Module, data):
        logits, features = net(data, return_feature=True)
        conf_scores, pred = torch.max(logits, dim=1)

        num_classes = net.fc.out_features

        class_score = []
        for mu_hat in self.mu_hat_per_class:
            features = features / torch.norm(features , dim = 1 , keepdim=True)
            score = self._phi(features) @ mu_hat
            class_score.append(score)

        class_score = torch.stack(class_score, dim=-1)

        conf_scores = torch.max(class_score , dim=-1).values

        return pred, conf_scores
```

```python
from tqdm import tqdm
import numpy as np
import torch
import torch.nn as nn
from openood.postprocessors import BasePostprocessor


class ClassConditionalRFFMeanPredictorScorePostprocessor(BasePostprocessor):
    """
    Class-Conditional Random Fourier Feature Mean Embedding OOD detector
    """

    def __init__(
        self,
        num_rff=4096,
        sigma=0.5,
        fit_ratio=0.95,
    ):
        super().__init__(None)

        self.D = num_rff
        self.sigma = sigma
        self.fit_ratio = fit_ratio

        self.mu_hat_per_class = []
        self.setup_flag = False

        self.APS_mode = False
        self.hyperparam_search_done = False



    def _sample_rff_params(self, feat_dim, device):
        self.omega = torch.randn(
            feat_dim, self.D, device=device
        ) / self.sigma

        self.b = 2 * np.pi * torch.rand(self.D, device=device)

    def _phi(self, x):
        """
        x: [N, feat_dim]
        return: [N, D]
        """
        projection = x @ self.omega + self.b
        return torch.sqrt(
            torch.tensor(2.0 / self.D, device=x.device)
        ) * torch.cos(projection)



    def setup(self, net: nn.Module, id_loader_dict, ood_loader_dict):

        if self.setup_flag:
            return

        print("\nEstimating Class-Conditional RFF mean embeddings...")

        net.eval()
        num_classes = net.fc.out_features
        self.mu_hat_per_class = [None] * num_classes

        all_feats = []
        all_labels = []

        with torch.no_grad():
            for batch in tqdm(id_loader_dict['train'], desc="calibrating class-conditional mean embeddings"):
                data = batch['data'].cuda()
                labels = batch['label'].cuda()
                _, features = net(data, return_feature=True)
                features = features / torch.norm(features , dim = 1 , keepdim=True)
                all_feats.append(features)
                all_labels.append(labels)

        all_feats = torch.cat(all_feats, dim=0)
        all_labels = torch.cat(all_labels, dim=0)
        N, feat_dim = all_feats.shape
        device = all_feats.device

        self._sample_rff_params(feat_dim, device)
        for c in range(num_classes):
            class_feats = all_feats[all_labels == c]
            if len(class_feats) > 0:
                phi_fit = self._phi(class_feats)
                self.mu_hat_per_class[c] = phi_fit.mean(dim=0)
            else:
                print(f"Warning: No training samples found for class {c}. Skipping mean embedding calculation for this class.")

        self.setup_flag = True
        print("Class-Conditional RFF mean embeddings estimated.")

    @torch.no_grad()
    def postprocess(self, net: nn.Module, data):
        logits, features = net(data, return_feature=True)
        model_conf_scores, pred = torch.max(logits, dim=1)

        num_classes = net.fc.out_features

        class_score = []
        for mu_hat in self.mu_hat_per_class:
            features = features / torch.norm(features , dim = 1 , keepdim=True)
            score = self._phi(features) @ mu_hat
            class_score.append(score)

        class_score = torch.stack(class_score, dim=-1)

        conf_scores = torch.max(class_score , dim=-1).values

        return pred, conf_scores * model_conf_scores
```

```python
import torch
import torch.nn as nn
import numpy as np
from tqdm import tqdm
from openood.postprocessors import BasePostprocessor

class KernelPCACoRPPostprocessor(BasePostprocessor):
    """
    Global RFF-based Kernel PCA (CoRP) Postprocessor.
    Does not store training features, only RFF parameters and global subspace components.


    reference : https://github.com/fanghenshaometeor/ood-kernel-pca/blob/main/CoP_CoRP_c10.py

    """
    def __init__(self, num_rff=4096, sigma=0.5, exp_var_ratio=0.99):
        super().__init__(None)
        self.D = num_rff
        self.sigma = sigma # Acts as gamma in the reference code
        self.exp_var_ratio = exp_var_ratio
        self.K = None

        self.setup_flag = False
        self.APS_mode = False
        self.hyperparam_search_done = False

        # Parameters to store
        self.omega = None
        self.b = None
        self.mu = None
        self.U = None

    def _sample_rff_params(self, feat_dim, device):
        # Matches: w = np.sqrt(2*gamma)*np.random.normal(size=(M,m))
        self.omega = torch.randn(feat_dim, self.D, device=device) * np.sqrt(2 * self.sigma)
        # Matches: u = 2 * np.pi * np.random.rand(M)
        self.b = 2 * np.pi * torch.rand(self.D, device=device)

    def _phi(self, x):
        # Matches: ftrain = np.sqrt(2/M)*np.cos((ftrain.dot(w.T)+u[np.newaxis,:]))
        projection = x @ self.omega + self.b
        return torch.sqrt(torch.tensor(2.0 / self.D, device=x.device)) * torch.cos(projection)

    def setup(self, net: nn.Module, id_loader_dict, ood_loader_dict):
        if self.setup_flag:
            return

        print("\nFitting Global CoRP (RFF PCA)...")
        net.eval()

        all_feats = []
        with torch.no_grad():
            for batch in tqdm(id_loader_dict['train'], desc="Collecting training features"):
                data = batch['data'].cuda()
                _, features = net(data, return_feature=True)
                # L2 normalize features exactly like normalizer in reference
                features = features / (torch.norm(features, p=2, dim=-1, keepdim=True) + 1e-10)
                all_feats.append(features)

        all_feats = torch.cat(all_feats, dim=0)
        feat_dim = all_feats.shape[1]
        device = all_feats.device

        self._sample_rff_params(feat_dim, device)

        # Map to RFF space
        Z = self._phi(all_feats)

        print("Fitting global PCA on RFF features...")
        # Matches: mu = ftrain.mean(axis=0)
        self.mu = Z.mean(dim=0, keepdim=True)
        Z_centered = Z - self.mu

        # Matches: K = ftrain.T.dot(ftrain)
        K_mat = Z_centered.T @ Z_centered

        # Matches: u_full, s, _ = np.linalg.svd(K)
        U_full, S, _ = torch.linalg.svd(K_mat)

        # Select q (or K) based on explained variance ratio
        s_accuml = torch.cumsum(S, dim=0) / torch.sum(S)
        valid_indices = torch.where(s_accuml >= self.exp_var_ratio)[0]
        if len(valid_indices) > 0:
            self.K = valid_indices[0].item() + 1
        else:
            self.K = S.shape[0]

        print(f"Explained variance ratio target: {self.exp_var_ratio}")
        print(f"Reduction dimension q (K) selected: {self.K}")

        # Top K components
        self.U = U_full[:, :self.K]

        self.setup_flag = True
        # Explicitly delete features to free memory
        del all_feats, Z, Z_centered, K_mat
        print("Global CoRP fitting complete. Training features have been discarded.")

    @torch.no_grad()
    def postprocess(self, net: nn.Module, data):
        logits, features = net(data, return_feature=True)
        conf, pred = torch.max(logits, dim=1)

        # L2 normalize features exactly like normalizer in reference
        features = features / (torch.norm(features, p=2, dim=-1, keepdim=True) + 1e-10)

        # Map to RFF space
        z = self._phi(features)

        # Center using global mean
        z_c = z - self.mu

        # Matches: reconstruct_in = u_q.dot(u_q.T).dot(ftest.T).T
        # mathematically equivalent to z_c @ U @ U^T
        recon = z_c @ self.U @ self.U.T

        # Matches: scores_in = - np.linalg.norm(ftest-reconstruct_in, ord=2, axis=1)
        recon_error = torch.norm(z_c - recon, p=2, dim=1)

        # Negative reconstruction error
        conf_scores = -recon_error

        return pred, conf_scores
```

```python
import pandas as pd
import itertools

def evaluate_multiple_configs(id_names, nets, postprocessors, data_root='./data', batch_size=200, num_workers=2):

    if len(id_names) != len(nets):
        raise ValueError("id_names and nets must have the same length.")

    all_metrics = []

    for (id_name, net), postprocessor in itertools.product(zip(id_names, nets), postprocessors):
        print(f"\nEvaluating with ID: {id_name}, Net: {net.__class__.__name__}, Postprocessor: {postprocessor.__class__.__name__}")

        # Initialize the evaluator with the current configuration
        evaluator = Evaluator(
            net,
            id_name=id_name,
            data_root=data_root,
            config_root=None,
            preprocessor=None,
            postprocessor=postprocessor,
            batch_size=batch_size,
            shuffle=False,
            num_workers=num_workers
        )

        # Perform OOD evaluation
        metrics = evaluator.eval_ood(fsood=False)

        # Assign the index to a new column named 'ood_dataset'
        metrics['ood_dataset'] = metrics.index.values
        metrics.index.name = None # Clear the index name if desired

        # Add metadata columns
        metrics['id_name'] = id_name
        metrics['postprocessor_name'] = postprocessor.__class__.__name__
        metrics['net_name'] = net.__class__.__name__

        all_metrics.append(metrics)

    # Combine all individual metrics DataFrames into one
    combined_metrics_df = pd.concat(all_metrics, ignore_index=True)
    return combined_metrics_df
```

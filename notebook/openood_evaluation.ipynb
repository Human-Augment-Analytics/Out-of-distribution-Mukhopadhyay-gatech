5. Extending OpenOOD for your own research/development

We try to make OpenOOD extensible and convenient for everyone.


You can evaluate your own trained model as long as it has necessary functions/methods that help it work with the postprocessors (see OpenOOD/openood/resnet18_32x32.py for example).


You can also design your own postprocessor by inheriting the base class (OpenOOD/openood/postprocessors/base_postprocessor.py), and the resulting method can be readily evaluated with OpenOOD.

# Custom Preprocessor using Kernel method

```python
from tqdm import tqdm
import numpy as np
import torch
import torch.nn as nn
from openood.postprocessors import BasePostprocessor
class RFFMeanPostprocessor(BasePostprocessor):
    """
    Random Fourier Feature Mean Embedding OOD detector
    Fully self-contained (no external YAML config needed)
    """

    def __init__(
        self,
        alpha=0.01,
        num_rff=512,
        sigma=1.0,
        fit_ratio=0.95,
    ):
        super().__init__(None)

        # Hyperparameters live inside class
        self.alpha = alpha
        self.D = num_rff
        self.sigma = sigma
        self.fit_ratio = fit_ratio

        self.setup_flag = False

        self.APS_mode = False
        self.hyperparam_search_done = False

    # --------------------------------------------------
    # Random Fourier Features
    # --------------------------------------------------

    def _sample_rff_params(self, feat_dim, device):
        self.omega = torch.randn(
            feat_dim, self.D, device=device
        ) / self.sigma

        self.b = 2 * np.pi * torch.rand(self.D, device=device)

    def _phi(self, x):
        """
        x: [N, feat_dim]
        return: [N, D]
        """
        projection = x @ self.omega + self.b
        return torch.sqrt(
            torch.tensor(2.0 / self.D, device=x.device)
        ) * torch.cos(projection)

    # --------------------------------------------------
    # Setup (Training Phase)
    # --------------------------------------------------

    def setup(self, net: nn.Module, id_loader_dict, ood_loader_dict):

        if self.setup_flag:
            return

        print("\nEstimating RFF mean embedding...")

        net.eval()
        all_feats = []

        with torch.no_grad():
            for batch in tqdm(id_loader_dict['train'] , desc="calibrating quantiles"):
                data = batch['data'].cuda()
                _, features = net(data, return_feature=True)
                all_feats.append(features)

        all_feats = torch.cat(all_feats, dim=0)
        N, feat_dim = all_feats.shape
        device = all_feats.device


        num_fit = int(self.fit_ratio * N)
        perm = torch.randperm(N, device=device)

        X_fit = all_feats[perm[:num_fit]]
        X_cal = all_feats[perm[num_fit:]]

        self._sample_rff_params(feat_dim, device)


        phi_fit = self._phi(X_fit)
        self.mu_hat = phi_fit.mean(dim=0)  # [D]

        phi_cal = self._phi(X_cal)
        scores = phi_cal @ self.mu_hat

        self.tau = torch.quantile(scores, self.alpha)

        print(f"Calibration threshold τ = {self.tau:.6f}")
        print(f"kernel :{self.omega}")

        self.setup_flag = True


    @torch.no_grad()
    def postprocess(self, net: nn.Module, data):

        net.eval()
        logits, features = net(data, return_feature=True)

        phi_x = self._phi(features)
        m_x = phi_x @ self.mu_hat

        # OpenOOD expects larger score = more ID-like
        conf = m_x

        pred = logits.argmax(dim=1)

        return pred, conf
```

```python
evaluator = Evaluator(
    net,
    id_name='cifar10',                     # the target ID dataset
    data_root='./data',                    # change if necessary
    config_root=None,                      # see notes above
    preprocessor=None,                     # default preprocessing for the target ID dataset
    postprocessor_name=None, # the postprocessor to use
    postprocessor=RFFMeanPostprocessor(),                    # if you want to use your own postprocessor
    batch_size=200,                        # for certain methods the results can be slightly affected by batch size
    shuffle=False,
    num_workers=2)                         # could use more num_workers outside colab
```

```python
metrics = evaluator.eval_ood(fsood=False)
```
